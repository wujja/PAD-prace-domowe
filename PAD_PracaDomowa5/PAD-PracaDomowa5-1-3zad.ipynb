```python
import pandas as pd
import numpy as np

df_orders = pd.read_csv("orders.csv", sep=",")
df_orders
```

```python
df_customers = pd.read_csv("customers.csv", sep=",")
df_customers
```

```python
#Zad 1a
df_orders["order_date"] = pd.to_datetime(df_orders["order_date"]).dt.date
df_orders
```

```python
#Zad 1b
df_orders["tshirt_category"].value_counts()
```

```python
#Zad 1c
def change_names_of_t_shirt(text):
    '''
    list_text = text.split()
    if('Wh' in list_text):
        list_text[0] = "White"
    elif('Bl' in list_text):
        list_text[0] = "Black"
    if('Tshirt' in list_text):
        list_text[1] = "T-Shirt"
    '''
    text = text.replace("Wh ", "White ")
    text = text.replace("Bl ", "Black ")
    text = text.replace("Tshirt", "T-Shirt")
    #return ' '.join(list_text)
    return text
```

```python
df_orders["tshirt_category"] = df_orders["tshirt_category"].apply(change_names_of_t_shirt)
df_orders["tshirt_category"].value_counts()
```

```python
#Zad 1d
def split_tshirt_category_by_color(text):
    list_text = text.split()
    if('Black' in list_text):
        return "Black"
    elif('White' in list_text):
        return "White"
    else:
        return pd.NA

def split_tshirt_category_by_size(text):
    list_text = text.split()
    if("M" in list_text):
        return "M"
    elif("F" in list_text):
        return "F"
    else:
        return pd.NA

def split_tshirt_category_by_type(text):
    list_text = text.split()
    if("T-Shirt" in list_text):
        return "T-Shirt"
    elif("Hoodie" in list_text):
        return "Hoodie"
    else:
        return "Tennis Shirt"

df_orders["tshirt_colour"] = df_orders["tshirt_category"].apply(split_tshirt_category_by_color)
df_orders["tshirt_size"] = df_orders["tshirt_category"].apply(split_tshirt_category_by_size)
df_orders["tshirt_type"] = df_orders["tshirt_category"].apply(split_tshirt_category_by_type)
df_orders
```

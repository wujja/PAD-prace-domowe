```python
import pandas as pd
import numpy as np

df_dane = pd.read_csv("PAD_03_PD.csv", sep=";", index_col='ID')

df_dane
```

```python
#Zadanie 1
df_dane["Country"].value_counts()
```

```python
#Zadanie 2
df_dane["owned_goods"] = df_dane["owns_car"] + df_dane["owns_TV"] + df_dane["owns_house"] + df_dane["owns_Phone"]
df_dane
```

```python
#Zadanie 3
mean_M = round(df_dane[df_dane["gender"] == "M"]["owned_goods"].mean(),2)
mean_K = round(df_dane[df_dane["gender"] == "K"]["owned_goods"].mean(),2)

print(f'Średnia posiadanych dóbr dla meżczyzn: {mean_M}, dla kobiet {mean_K}')
```

```python
#Zadanie 4
df_groupby = df_dane.groupby(['Country'])["Country","owned_goods","Age"].agg({'owned_goods':'mean','Age':'min'})
df_groupby['owned_goods'] = round(df_groupby['owned_goods'],2)
df_groupby
```

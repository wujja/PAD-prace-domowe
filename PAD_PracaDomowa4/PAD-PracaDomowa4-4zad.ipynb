```python
import pandas as pd
import numpy as np



data = pd.read_csv("Zadanie_4.csv", sep=";")
data["Dates"] = pd.to_datetime(data['DateTime']).dt.date
data["Times"] = pd.to_datetime(data["DateTime"]).dt.time
data.dropna(inplace=True, axis=1)
data
```

```python
data_minData = data.groupby(['DoctorID','Dates',]).agg({'Times':'min'})
data_minData.rename(columns={'Times':'Start'},inplace=True)
data_maxData = data.groupby(['DoctorID','Dates']).agg({'Times':'max'})
data_maxData.rename(columns={'Times':'End'},inplace=True)
```

```python
data_result = pd.concat([data_minData,data_maxData], axis=1, join='inner')
data_result
```
